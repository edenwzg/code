# 蹦床游戏 + 墙面双拼-儿童版-产品设计案

> 版本: 1.1.20171221  
> 编辑者: 王泽光

## 注意事项

- 产品面向的用户为儿童 __10__ 岁以下
- 产品承载的人数不能太少(最好是没有硬性要求)
- 游戏要可随时进出,满足开放式运营,无人化运营
- 游戏内容设计,避免强迫用户频繁(<2秒)的在两个画面交替观察
- 在 __240*180__ 的蹦床上 __>3__ 人的娱乐环境,将会大大降低游戏体验, __>6__ 人, __kinext__ 将会无法识别多余的玩家
- 游戏内容的设计重点是 __简洁的趣味性__ ,以弥补游戏玩法不足而产生的枯燥感
- 每个玩家的有效识别在画面中都要有 __反馈__ ,特别是在人数较多的情况下也要有所表现.
- 所有游戏内容的设计都必须能够符合统一的游戏模式.
- 所有数值都要做成可配置的形式(不要写死)
- 产品可以离线运行.
- 联网后游戏可以 __在线更新__ .更新后只需重新启动游戏,无需其他额外的操作.
- 产品为 __系列产品__ ,游戏内容不能设计的过于复杂.
- 多人混乱游戏模式,不好设计需要让玩家根据某个短暂时机做出相应的行为
- 蹦床上不好设计让玩家根据某个短暂时机做出跳跃响应的行为
- 多人混乱游戏模式,不好精确控制某个角色.
- 蹦床上不好设计让玩家边跳边做手势的设计(尤其是不限制人数的产品)
- 双拼产品在设计和实现上有一定难度和内容开发成本,可能要考虑另外的商业模式(区别于原有系列游戏的商业模式)比如软件收费.
- 安全问题:
    - 当投影机安装过低时,会引发安全问题,游戏内跳跃值会鼓励玩家跳跃,放大安全问题
    - 地面和蹦床上,因人为导致的震动\晃动,不能传递到投影仪的固定位置
    - 避免设计需要玩家前后跳跃的玩法,有可能玩家注意力在墙面,在前后跳跃时从蹦床后面摔下来.
- 应用场景和收费模式
- 商家活动：免费或商家组织
- 蹦床公园：套票统一计时
- 主题场馆：套票统一计时
- 健身会所：整馆按次或会员卡
- 成人蹦床的游戏适合单人单床，不适合多人一床（成人比较介意相互之间的影响）
- 成人玩蹦床游戏的时候手机一般不在身上
- 单独设备可独立体验游戏内容也可以联网PK
- 要根据玩家的跳跃高度获取高度数据


## 产品设计

### 成人蹦床

- 单/双人蹦床（>1000￥） + 投影 + Kinext2.0（>2000￥）
- 场地高度要求4米以上
- 蹦床面离地高度比儿童版的要高。
- 投影画面的尺寸240cm*320cm左右
- 一个画面可以支持一组双人蹦床每个蹦床区域为边长120cm左右的正方形（不加护垫）
- 护垫宽度为60cm

### 儿童蹦床

展厅环境和现有一体机尺寸
- 展厅顶棚高度： 300cm
- 一体机机身深度： 40cm

蹦床
- 蹦床床面尺寸为： 240cmx180cm ，蹦床高度为 50cm ，软垫宽度为 40cm
- 蹦床较宽的一面靠近墙面安放（如果时正方形蹦床则任意边靠近墙面）

墙面投影一体机
- 墙面投影画面尺寸： 240cm*180cm ，宽高比为： 4：3（0。75）
- 墙面一体机 不加装kinext
- 墙面一体机距离墙面的距离为： 260cm （一体机挂架点）
- 墙面投影仪挂在蹦床距离画面较远的一侧，略低于蹦床投影一体机，保证画面不会被前方的机箱遮挡

场地
- 场地高度最小为 300cm 

### 健身蹦床

- 便携蹦床（<300￥） + 电视 + 投影 + 姿态测量传感器（<100￥）
- 可针对家庭扩展销售渠道
- 可针对小型专业场地扩展销售渠道（如健身房，康复中心，早教中心）
- 让读取用户ID成为更合理的，并通过ID做客户深度开发（如购买道具）
- 可做无人化运营

## 游戏模式

- 游戏以局（inning）为一个玩法单位
- 游戏规则为：在规定的时间内，完成某件事情， 或则在规定的时间内， 看谁（或游戏中的替身）完成的最好。在内容设计上不突出个人
- 每局游戏没有失败条件（no losers）。但是游戏内容结束后，会根据游戏内容完成的情况打出 评价（rate）
- 每局游戏分为3个阶段（stage）：准备阶段，游戏阶段，评价阶段。开始阶段和评价阶段，在形式设计上采用统一的模式，便于玩家理解
- 开始阶段让玩家准备开始游戏了解接下来的游戏内容（或猜测游戏内容玩法），让玩家意识到新的一局即将开始，开始阶段的时间控制在 3 秒左右
- 游戏阶段提供游戏内容 每次随机读取游戏内容，或顺序读取游戏内容。（可在配置文件中设置）随机读取的内容不能与前一次的重复
- 评价阶段对玩家展示游戏结果 对该局游戏谁（或游戏中的替身）获胜，或完成到什么程度给予评价。表现浅显易懂，简单明了。结束阶段的时间控制在 5 秒左右
- 每个游戏内容都会限定 时间（time），根据游戏内容的不同时间也会不同。 时间结束前三秒会进入倒计时，时间结束后会进入评价阶段
- 每个游戏内容都有一个浅显易懂的目标（target），大部分游戏参与人数越多，目标达成越快

### 机甲守卫

问题:

- ~~炮弹筒没有必要做在蹦床面上,不利于玩家的观察感受~~

设计
- 本游戏为一款横板塔防游戏.
- 画面风格为卡通
- 机器人从画面左侧飞出来,飞向画面右侧,机器人有生命值
- 游戏的目标是尽量击毁所有机器人
- 画面下方为5个炮筒,玩家再蹦床的不同位置跳跃,对应的炮筒会发射炮弹
- 根据威力不同,炮弹可分为不同的类型,在外观上也有差别,同时对机器人也会造成不同的伤害
- 机器人被击中后会减血,血减为0会爆炸,爆炸后会散落一些零件在地面上.零件一定时间后会消失
- 根据玩家的跳跃高度不同会产生不同威力的炮弹
- 游戏结束后会根据击毁的机器人给出本局游戏的评价
- 这是一个多人配合的游戏,人数越多,拿到高评价的几率会大大增加

### 跳跳球

问题
- ~~通过让球飞出来后,在蹦床上踩球结束游戏规则比较隐晦,也容易脱离游戏的打转玩法.~~
- ~~蹦床内容没有实际的玩法价值(去掉蹦床画面的玩法)~~

设计
- 本游戏是一个打砖块游戏，画面中由一些可破坏的砖块组成
- 游戏由一个封闭的空间和空点顶部的一排可破坏的方块墙,以及一个带表情,带小手的跳跳球组成
- 画面底部是一个虚拟的弹力波,玩家跳跃会引发弹力波的波动
- 弹力波有两个作用,一是击中跳跳球后会给跳跳球增加弹力,二是改变跳跳球方向移动轨迹
- 如果弹力波没有击中跳跳球,那么跳跳球的弹力会衰减
- 玩家的手部动作,会改变跳跳球的表情和小手的动作(预定义的动作),跳跳球落地后有变形效果
- 调调球在碰到墙壁时会有反弹效果,同时也会衰减弹力度
- 跳跳球飞到高处的目的是将高处的方块击碎,(类似打砖游戏)
- 跳跳球在击碎特殊方块后会分切为多个跳跳球.
- 砖块按照坚固程度可分为不同的类型
- 在规定时间内将砖块击碎获胜
- 这个模式可以单独拿出来进一步优化玩法,做成多关卡的独立游戏模式

### 卫星发射

适用于
- 成人蹦床
- 儿童蹦床
- 健身蹦床

问题
- 火箭游戏缺乏游戏深度
- 在玩家跳跃的过程中，考虑画面反馈

设计
- 游戏的场景是一个卫星发射基地，玩家的目的是吧火箭升空，并在规定的时间内把卫星送入轨道
- 火箭有三个发动机（火箭推进器）不同位置的玩家跳跃会影响不同位置的发动机
- 玩家的跳跃高度影响火箭飞行的速度和火焰特效的大小
- 火箭初始离地需要很大的推力，玩家需要努力的跳跃
- 当火箭升空后，玩家仍然需要不断的给火箭加速，否则无法在规定的时间内，脱离大气层
- 不同位置的发动机决定了火箭的飞行角度，玩家要相互配合保持两边发动机的推力不要相差太大
- 当火箭倾斜时要通过两侧的火箭发动机及时调整。否则当火箭角度过大时将会解体
- 当火箭倾斜超过一定角度时
- 火箭进入大气层会发生震动，并加剧火箭的不稳定性。同时火箭顶端会因摩擦产生火焰
- 火箭进入太空后，会抛出卫星进入轨道。玩家圆满完成了任务
- 如果没有在规定时间内进入将卫星送达目标，火箭燃料将耗尽，并坠落爆炸
- 游戏结束后评价此次发射总共用了多少时间


实现
- 蹦床上的每个点包含两个值[R，L]
- 玩家跳跃的高度为D
- 推力S = (R-L).abs*D
- (R-L)为倾斜角度，方向为[R，L].max
- 玩家每次跳跃都要进行累计计算，并根据结果进行状态判断

### 爆炸糖

问题
- ~~游戏的目的是通过吃糖将小朋友的体力消耗光,不利于玩家理解~~
- ~~要表述的内容不直接:吃糖游戏传达的是要少吃糖~~

设计
- 这是一个绚丽的五彩缤纷的游戏
- 画面中是一个张着大嘴的卡通形象,样子有点调皮,有一个爆炸头型,和一排整齐的牙齿
- 卡通形象旁边有两只黏黏的小手,和两个转动的大眼睛
- 在蹦床上没有玩家的时候,它会偶尔往嘴里扔糖,并做出高兴夸张的表情
- 蹦床画面中是各种糖果
- 当玩家跳跃时,墙面会飞入各种糖果(包括爆炸糖),卡通形象看到糖后会用手吧糖抓向或拍向嘴里
- 他会根据吃到的糖果类型,做出各种夸张表情
- 卡通角色在吃糖果的过程中,牙齿会不断脱落
- 游戏的目的是在规定的时间内通过吃糖让他的牙齿脱落
- 时间结束后的评价是通过这个小朋友会掉落多少牙齿给出评价
- 这个游戏的目的是提醒小朋友不要吃太多的糖果

### 海底套圈

- 墙面是一个3D的海底世界,有各种海底生物和鱼类,海底有各种不同颜色的皮圈
- 玩家在蹦床上跳跃,海底床上会吹出气泡,如果气泡上面有皮圈,它将会被吹起来
- 皮圈在上升的过程中会不断地缓慢的自身旋转,如果没有气泡的推力,它将会沉到海底.
- 皮圈只会在一个2维区域内移动,不会沿摄像机深度方向移动.
- 当橡皮圈上升到一定高度,并且鱼与皮圈的距离比较近时,鱼会加速从橡皮圈中穿越
- 当鱼从橡皮圈中穿越后,橡皮圈会转换为一个图标,飞向画面上方的计分区,并累加皮圈数
- 当皮圈消失后,海底会刷新一个新的皮圈
- 游戏时间结束后,根据获得的皮圈数量给出这局游戏的评价

### 水柱打昆虫


- 画面场景是一个草坪, 草坪上有一些喷泉口, 画面中间飞着很多昆虫.
- 玩家在蹦床上的每一次跳跃,会引发喷水口喷水
- 水柱的高度和方向与玩家的跳跃深度和位置有关
- 当水柱击中昆虫后,会导致其飞行高度降低或直接落地.
- 游戏时间结束后会根据最终击落的昆虫数量给出本局游戏的评价

### 一飞冲天

- 每名玩家控制单个角色
- 角色背着火箭背包的，最初站在地面上
- 玩家起跳的过程会触发游戏内的火箭背包喷射火焰，给游戏中的角色增加向上的推力
- 游戏目标：在规定的时间内（30秒），争取飞的更高
- 玩家的跳跃高度，会反应到游戏中角色的火箭背包喷射火焰的强度。
- 游戏画面向上无限延伸，在飞行过程中会经过 地面，摩天大楼，鸟群，飞机，云层，太空
- 玩家在跳跃的过程中会遭遇障碍物，鸟群，飞机等，玩家需要控制跳跃力度避免撞上障碍物。
- 在出现障碍的情况下，玩家需要控制跳跃力度避免游戏中的角色不触碰到障碍。

### 障碍马术大赛

- 硬件需求：单人小蹦床 + 姿态测量传感器（或Kinect）
- 针对希望通过游戏的方式做有氧运动锻炼身体的用户
- 通过提供高强度，不同频率的跳跃形式来传达游戏运动健身理念
- 游戏形式通过障碍马术的形式包装，贴合运动行为。
- 游戏场景为马术障碍赛场， 游戏为第三人称视角
- 游戏可以支持双人或三人分屏。
- 赛马会在场地内按照随机路线奔跑
- 在无障碍的情况下，玩家的跳跃会增加马奔跑的速度
- 在跳跃障碍之前（通过triger控制）玩家的跳跃会触发马的跳跃
- 玩家跳跃给马加速的目的有两个
- 第一：有些比较宽的障碍需要一定的速度才能越过
- 第二：如果马速太快，马屁有拒跳的可能
- 第二：完成比赛的用时会计入总分
- 如果马触碰障碍不会被绊倒，但会影响速度。
- 比赛规则
- 满分为0分，如马匹在越过障碍物时碰倒障碍物就会被罚4分
- 马匹拒跳，扣除8分。


### 滑雪大冒险

- 模仿手机游戏<滑雪大冒险>
- 单人技巧挑战类游戏
- 玩家可以通过各种控制体态控制游戏中的角色，产生不同的滑雪效果
- 针对不同层面的玩家（普通玩家和深度职业玩家）
